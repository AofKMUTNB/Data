```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srs

x = np.linspace(2,20,10)
x

plt.plot(x, x)
plt.show()
```

```python
plt.plot(x,x)
plt.plot(x,x**2)
plt.show
```

```python
plt.plot(x,x,'k-D')
plt.plot(x,x**2)
plt.xlabel('value')
plt.ylabel('Day')
plt.show
```

```python
plt.plot(x,x,'k-D',label = 'y = x')
plt.plot(x,x**2,label = 'y = x**2') 
plt.xlabel('value')
plt.ylabel('Day')
plt.legend()
plt.show
```
