```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srs

x = np.linspace(2,20,10)
x

plt.plot(x, x)
plt.show()
```

```python
plt.plot(x,x)
plt.plot(x,x**2)
plt.show
```

```python
plt.plot(x,x,'k-D')
plt.plot(x,x**2)
plt.xlabel('value')
plt.ylabel('Day')
plt.show
```

```python
plt.plot(x,x,'k-D',label = 'y = x')
plt.plot(x,x**2,label = 'y = x**2') 
plt.xlabel('value')
plt.ylabel('Day')
plt.legend()
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srs
x = ['Shirt','Pants','Shorts','Shoes']
y = [1000,1200,800,200]

plt.bar(x,y)
plt.title('Sale')
plt.xlabel('Catagory')
plt.ylabel('Price')
plt.show()
```

```python
plt.barh(x,y)
plt.title('Sale')
plt.xlabel('Catagory')
plt.ylabel('Price')
plt.show()
```

```python
x = np.random.normal(size=100)
```

```python
fig, axes = plt.subplots(nrows=3,ncols=2,figsize=(8,8))

axes[0,0].set_title('Demo')
axes[0,0].plot(np.linspace(1,10,10),np.linspace(1,10,10),'k-D',label = 'y = x')
axes[0,0].plot(np.linspace(1,10,10),np.linspace(1,10,10)**2,label = 'y = x**2') 
axes[0,0].set_xlabel('value')
axes[0,0].set_ylabel('Day')
axes[0,0].legend()

axes[1,1].set_title('Demo1')
axes[1,1].plot(np.linspace(1,10,10),np.linspace(1,10,10),'k-D',label = 'y = x')
axes[1,1].plot(np.linspace(1,10,10),np.linspace(1,10,10)**2,label = 'y = x**2') 
axes[1,1].set_xlabel('value')
axes[1,1].set_ylabel('Day')
axes[1,1].legend()
```
